# Exploratory Data Analysis on Training Data

```python
data = pd.read_csv('datasets/final_preprocessed.csv')

#remove first two 'index' columns
data = data.iloc[: , 2:]

data
```

```python
for col in data.columns:
    print(col, data[col].isnull().sum())
```

```python
datana = data.dropna()
datana
```

```python
datana['Star'] =  datana.Stars.str.split(' ', expand = True)[1]
#datana['Stars'] = datana.Stars.astype(float)
datana['Rating'] = pd.to_numeric(datana.Star, errors='coerce').astype(np.float)
datana
```

```python
#r = datana.Restaurant.unique()
#restslist = datana.sort_values(r, ascending=True)
avg = round(datana.groupby('Restaurant')['Rating'].mean(), 4)
avg = pd.DataFrame(avg)
#avg["Restaurants Name"] = avg.iloc[: , 0:1]
#avg["Average Rating"] = avg.iloc[: , 1:]
#average = pd.DataFrame()

avg["Restaurant Name"] = avg.index
avg["Average Rating"] = avg.Rating
avg = avg.iloc[2: , :]

avg
```

```python
def show_values(axs, orient="v", space=.02):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.3f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.2)
                value = '{:.3f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)

import matplotlib.pyplot as plt
fig = plt.gcf()

# Change seaborn plot size
fig.set_size_inches(12, 8)
#create horizontal barplot
p = sns.barplot(x="Average Rating", y="Restaurant Name", data=avg, ci=None)
sns.set(font_scale=1.5)
#show values on barplot
show_values(p, "h", space=0)
```

# Exploratory Data Analysis on Prediction Data

```python
import pickle
path= 'sent_datasets/reg_sent_df.pkl' #path=â€™/root/â€¦â€¦/aus_openface. pklâ€™ Path where the pkl file is located.
f=open(path,'rb')
data2=pickle. load(f)
print(data2)
```

```python
for col in data2.columns:
    print(col, data2[col].isnull()
          .sum())
```

```python
data2['Star'] =  data2.stars.str.split(' ', expand = True)[1]
data2['Rating'] = pd.to_numeric(data2.Star, errors='coerce').astype(float)

avg2 = round(data2.groupby('restaurant')['Rating'].mean(), 4)
avg2 = pd.DataFrame(avg2)

avg2["Restaurant Name"] = avg2.index
avg2["Average Rating"] = avg2.Rating


fig2 = plt.gcf()

# Change seaborn plot size
fig2.set_size_inches(14, 8)
#create horizontal barplot
p2 = sns.barplot(x="Average Rating", y="Restaurant Name", data=avg2, ci=None)
sns.set(font_scale=1)
#show values on barplot
show_values(p2, "h", space=0)
```

```python
#average clause polarity by restaurant
cavg = round(data2.groupby('restaurant')['clause_polarity'].mean(), 4)
cavg = pd.DataFrame(cavg)
cavg["Restaurant Name"] = cavg.index
cavg["Average Clause Polarity"] = cavg.clause_polarity
cavg


fig2 = plt.gcf()

# Change seaborn plot size
fig2.set_size_inches(12, 8)
#create horizontal barplot
p2 = sns.barplot(x="Average Clause Polarity", y="Restaurant Name", data=cavg, ci=None)
sns.set(font_scale=1)
#show values on barplot
show_values(p2, "h", space=0)
```

```python
#average clause polarity by restaurant by category
ccavg = data2.groupby(['restaurant', 'category'])['clause_polarity'].mean().reset_index()
#ccavg = pd.DataFrame({'clause_polarity' : data2.groupby( [ 'restaurant', 'category'], as_index=False).mean()}).reset_index()
ccavg = pd.DataFrame(ccavg)

ccavg.head(10)
```

```python

sns.set_context('talk')
sns.set(font_scale=2)

# Create a grouped bar chart
fig, ax = plt.subplots(figsize=(35, 15))

# Our x-axis. 
x = np.arange(5) 

# Define bar width. We need this to offset the second bar.
bar_width = 0.1

b1 = ax.bar(x, ccavg.loc[ccavg['restaurant'] == 'Adda', 'clause_polarity'],
            width=bar_width, label='Adda')
# Same thing, but offset the x.
b2 = ax.bar(x + bar_width, ccavg.loc[ccavg['restaurant'] == 'AlAmeen', 'clause_polarity'],
            width=bar_width, label='Al Ameen')
b3 = ax.bar(x + bar_width + bar_width, ccavg.loc[ccavg['restaurant'] == 'Brio', 'clause_polarity'],
            width=bar_width, label='Brio')
b4 = ax.bar(x + bar_width + bar_width + bar_width, ccavg.loc[ccavg['restaurant'] == 'Comida_Mexicana', 'clause_polarity'],
            width=bar_width, label='Comida_Mexicana')
b5 = ax.bar(x + bar_width + bar_width + bar_width + bar_width, ccavg.loc[ccavg['restaurant'] == 'DTF', 'clause_polarity'],
            width=bar_width, label='DTF')

# Fix the x-axes.
ax.set_xticks(x + 0.4 / 2)
ax.set_xticklabels(ccavg.category.unique())

# Add legend.
ax.legend()

# Axis styling.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

# Add axis and chart labels.
ax.set_xlabel('Categories', labelpad=15)
ax.set_ylabel('Average Clause Polarity', labelpad=15)
ax.set_title('Some Restaurants by Category using Average Clause Polarity', pad=15)

fig.tight_layout()
```
